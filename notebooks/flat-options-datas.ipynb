```python
import os
from datetime import timedelta
import pandas as pd
from google.cloud import storage
from io import StringIO
```

```python
storage_client = storage.Client()
bucket_name = os.environ["BUCKET_NAME"]
bucket = storage_client.bucket(bucket_name)
```

```python
blobs = bucket.list_blobs()
expiration = '2024-03-15'
dataframes = {}
for blob in blobs:
    if "reports/" in blob.name:
        print(blob.name)
        print(blob.generate_signed_url( version="v2", expiration=timedelta(days=1), method="GET"))        
```

```python
blobs = bucket.list_blobs()
expiration = '2024-03-15'
dataframes = {}
for blob in blobs:
    if "chains/" in blob.name and "TLT" in blob.name:
        bytes_data = bucket.blob(blob.name).download_as_bytes()
        data = StringIO(str(bytes_data,'utf-8'))
        dataframe = pd.read_csv(data)
        if expiration:
            dataframes[blob.name] = dataframe[dataframe['expiration'] == expiration]
        else:
            dataframes[blob.name] = dataframe
```

```python
previes_key = None
new_dataframe = None
for key in dataframes:
    if new_dataframe is None:
        new_dataframe = dataframes[key]
    new_dataframe[key] = dataframes[key]['openInterest']
```
